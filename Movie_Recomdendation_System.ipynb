## **Movie Recommendation System**

**Recommender System** is a system that predict or filter preferences according to the user's choices. Recommender shystems are utilized in a variety of areas including movies, musics, news, books, research articles and products in general.
Recommender system produce a list of recommendations in any of the tow ways -

**Collaborative filltering:** Collaborative filltering approches build a model from the user's past behaviour as well as similar decisions made by other users. The model i then used to predict items that user may have an iterest in.

**Content-based filltering:** Content based filltering approaches uses a series of discrete charecteristics of an item in order to recommend additional items with similar properties.

# Objective


The main objective of building a Movie recommendation system is to provide the user an ultimate experience of using the aapp/website for watchiing movies based on his/her interest.

# Feature Selection

```python
data1=data.drop('Movie_ID',axis=1)
```

```python
new_df=data1[['Movie_Title', 'Movie_Genre', 'Movie_Language','Movie_Keywords', 'Movie_Overview',
              'Movie_Production_Country',
              'Movie_Spoken_Language', 'Movie_Tagline', 'Movie_Cast', 'Movie_Crew',
              'Movie_Director']].fillna('')
```

```python
new_df.shape
```

```python
new_df.count()
```

```python
x=new_df['Movie_Title']+' '+new_df['Movie_Genre']+' '+new_df['Movie_Language']+' '+new_df['Movie_Keywords']+' '+new_df['Movie_Overview']+' '+new_df['Movie_Production_Country']+' '+new_df['Movie_Spoken_Language']+' '+new_df['Movie_Tagline']+' '+new_df['Movie_Cast']+' '+new_df['Movie_Crew']+' '+new_df['Movie_Director']
```

```python
x.shape
```

# Get Feature Text Conversion to Tokens

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfid=TfidfVectorizer()
```

```python
x=tfid.fit_transform(x)
```

```python
x.shape
```

```python
# print(x)
```

# Get Similarity Score using Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(x)
```

```python
# similarity_score
```

```python
similarity_score.shape
```

# Get Movie name as input from User and Validate for Closest Spelling

```python
movie_name=input('Enter your favourite movie name: ')
```

```python
All_movies_list=data['Movie_Title'].tolist()
```

```python
import difflib
```

```python
movie_recomendation=difflib.get_close_matches(movie_name,All_movies_list)
print(movie_recomendation)
```

```python
close_match=movie_recomendation[0]
print(close_match)
```

```python
index_of_close_match=data[data.Movie_Title==close_match]['Movie_ID'].values[0]
print(index_of_close_match)
```

```python
#getting a list of similar movies
Recomendation_Score=list(enumerate(similarity_score[index_of_close_match]))

print(Recomendation_Score)
```

```python
#get Sorted movie list
sorted_movie_list=sorted(Recomendation_Score, key = lambda x:x[1],reverse=True)

print(sorted_movie_list)
```

```python
#get top 10 similar movie

print("Top 10 movies for you : \n")

i=1
for movies in sorted_movie_list:
  index=movies[0]
  title=data[data.Movie_ID==index]['Movie_Title'].values
  if(i<11):
    print(i,"-",title)
    i+=1
```


# Final Movie Recommendation System 

```python
movie_name=input("Enter your favourite movie name: ")

list_of_all_movies=data['Movie_Title'].tolist()

matching_movies=difflib.get_close_matches(movie_name,list_of_all_movies)

close_match=matching_movies[0]
# print(close_match)

Index_of_movie=data[data.Movie_Title==close_match]['Movie_ID'].values[0]

recomendation_score=list(enumerate(similarity_score[Index_of_movie]))

sorted_movie_list=sorted(recomendation_score,key=lambda x:x[1],reverse=True)

print("\nTop 10 movies for you :\n")

i=1

for movies in sorted_movie_list:
  index=movies[0]
  title=data[data.Movie_ID==index]['Movie_Title'].values
  link=data[data.Movie_ID==index]['Movie_Homepage'].values

  if(i<11):
    print(i,"-",title," Link -",link)
    i += 1
    
```
